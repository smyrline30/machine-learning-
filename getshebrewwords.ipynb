```python
from google.colab import files

uploaded = files.upload()  # upload multiple XML files
```

```python
import xml.etree.ElementTree as ET
import re

# Hebrew Unicode range
hebrew_pattern = re.compile(r'[\u0590-\u05FF]+')

def extract_hebrew_from_xml(file_path):
    words = []

    tree = ET.parse(file_path)
    root = tree.getroot()

    for elem in root.iter():
        if elem.text:
            matches = hebrew_pattern.findall(elem.text)
            words.extend(matches)

    return words
```

```python
all_words = []

for file_name in uploaded.keys():
    print(f"Processing: {file_name}")
    words = extract_hebrew_from_xml(file_name)
    all_words.extend(words)
```

```python
# Remove duplicates
unique_words = list(set(all_words))

print(f"Total words extracted: {len(unique_words)}")
```

```python
import pandas as pd

df = pd.DataFrame(unique_words, columns=["Hebrew Words"])

df.to_excel("hebrew_words.xlsx", index=False)
```
